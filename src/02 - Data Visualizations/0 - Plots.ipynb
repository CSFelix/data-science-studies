````python
import pandas as pd # pip install pandas
import matplotlib.pyplot as plt # pip install matplotlib
import seaborn as sns # pip install seaborn
import numpy as np # pip install numpy

np.random.seed(6082023)

sns.set_style('whitegrid')
````

````python
# ---- Datasets ----
spotify_df = pd.read_csv('./datasets/spotify.csv', index_col='Date', parse_dates=True)
flight_df = pd.read_csv('./datasets/flight_delays.csv', index_col='Month')
insurance_df = pd.read_csv('./datasets/insurance.csv')
iris_df = pd.read_csv('./datasets/iris.csv', index_col='Id')

# ---- Fakes Datasets ----
#
# Candidates DF:
#    - Year: year from 2021 to 2023
#    - Candidate: candidate's name
#    - Percentage Population: % of population that voted on the candidate
#
candidates_df = pd.DataFrame({
    'year'                    :  [2021, 2022, 2023] * 3
    , 'candidate'             :  np.repeat(['candidate_1', 'candidate_2', 'candidate_3'], 3)
    , 'percentage_population' :  [30, 25, 40, 40, 45, 35, 30, 30, 25]
})
````

<p id='00-line-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>0.0 | Line Plots</p>

> **Usage** - `numerical feature over a time feature. Can have a categorical feature as hue - seaborn does it automatically`.

`Line Plots` are useful to show the relationship between a `numerical and a time feature`. The numerical feature can be continuous or discrete, whatever.

About the time feature, it can be secons, minutes, hours, days, weeks, months, years, and so on...

> Plot Using All Features

```python
sns.lineplot(data=df)
```

> Plot Using A Few Features

```python
sns.lineplot(data=df['Feature1'], label='Label1')
sns.lineplot(data=df['Feature2'], label='Label2')
```

````python
# ---- Line Plot Using All Features ----
sns.lineplot(data=spotify_df)

plt.title('Songs Views: January-2017 to January-2018')
plt.ylabel('Views (per 10 Millions)')
plt.show()
````

````python
# ---- Line Plot Using A Few Features ----
sns.lineplot(data=spotify_df['Despacito'], label='Despacito')
sns.lineplot(data=spotify_df['HUMBLE.'], label='HUMBLE')
sns.lineplot(data=spotify_df['Unforgettable'], label='Unforgettable')

plt.title('Song Views: January-2017 to January-2018')
plt.ylabel('Views (per 10 Millions)')
plt.show()
````

<p id='01-point-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>0.1 | Point Plots</p>

> **Usage** - `numerical feature over a time feature. Can also have a categorical feature as hue`.

`Point Plots` are used like Line Plots, that is, to show a `numerical feature` over a `time feature`. The major difference between them is that this one is used when there are many observations, whereas that one is used when there are a few observations.

> Plot Without Categorical Hue

```python
sns.pointplot(data=df, x='Feature1', y='Feature2')
```

> Plot With Categorical Hue

```python
sns.pointplot(data=df, x='Feature1', y='Feature2', hue='Categorical_Feature')
```

````python
# ---- Point Plot Without Categorical Hue ----
#
# - realize that without putting a categorical hue, it's better
# to plot datas from only one categorical value.
# - dontcha believe me? Try yourself updating the 'data' parameter
# like this: data=candidates_df
#
sns.pointplot(
    data=candidates_df.loc[candidates_df['candidate'] == 'candidate_1']
    , x='year', y='percentage_population'
)

plt.title('Percentage Population Votes to Candidate 1 over the Years')
plt.show()
````

````python
# ---- Point Plot With Categorical Hue ----
#
# - when we set a Categorical Hue, there's no problem in showing
# datas from all categorical values
#
sns.pointplot(
    data=candidates_df
    , x='year', y='percentage_population', hue='candidate'
)

plt.title('Percentage Population Votes per Candidate over the Years')
plt.show()
````

<p id='10-box-and-whisker-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>1.0 | Box and Whisker Plots</p>

> **Usage** - `nuimerical feature distribution`.

`Box and Whisker Plots` are powerfull plots that shows the distribution of numerical features. In the distribution, we can see the `quartiles`, the `median (second quartile)`, the `lowest and highest observed values`.

Some times, it is even possible to see the `minimum and maximum values` for this feature, and `outliers`.

> **Quartiles** - `are represented by lines - whiskers - over the plot`;

> **Median** - `corresponds to the second quartile, that is, the second whisker`;

> **Lowest and Highest Observed Values** - `correspond to the first and fourth quartile simultaneously`;

> **Minimum and Maximum Values** - `correspond to the first and fourth quartile simultaneously when there are outliers. The minimum value is present when there are outliers smaller than it. The maximum value is present when  there are outliers higher than it`;

> **Outliers** - `are represented by points outside of the box`.

---

> Plot Without Categorical Hue

```python
# y feature is optional
sns.boxplot(data=df, x='Feature1', y='Feature2')
```

> Plot With Categorical Hue

```python
sns.boxplot(data=df, x='Feature1', y='Feature2', hue='Categorical_Feature')
```

````python
# ---- Box Plot Without Categorical Hue ----
sns.boxplot(data=spotify_df, x='Shape of You')

plt.title('View to "Shape of You" Song - January-2017 to January-2018')
plt.xlabel('Views (1.0 corresponds to 10 Millions)')
plt.show()
````

````python
# ---- Box Plot With Categorical Hue ----
#
# - setting 'candidate' as x-axis and hue allows seaborn to show a
# legend and to shrink large stretched boxes
#
#
sns.boxplot(
    data=candidates_df
    , x='candidate'
    , y='percentage_population'
    , hue='candidate'
)

plt.title('Percentage Population Votes per Candidate')
plt.show()
````

<p id='11-swarm-and-strip-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>1.1 | Swarm and Strip Plots</p>

> **Usage** - `categorical feature distribution over a numerical feature`.

`Swarm Plots` are used to show a categorical feature's distribution over a numerical one.

`Strip Plots` have the same usage of Swarm Plots, the differences ar that Strip Plots are used when the Swarm cannot show all points; also, Strip ones agglomerate the points, that is, some points can occupy the same space - Einstein would not like this!.

> Plot

```python
sns.swarmplot(data=df, x='Categorical_Feature', y='Numerical_Feature', hue='Categorical_Feature')

sns.stripplot(data=df, x='Categorical_Feature', y='Numerical_Feature', hue='Categorical_Feature')
```

````python
# ---- Swarm Plot ----
#
# - realize that are so many datas that will not be possible to plot
# them all and python will tell us to use strip plot instead
#
sns.swarmplot(data=insurance_df, x='smoker', y='charges', hue='smoker')

plt.title('Charges by Smoker or Non-Smoker People')
plt.show()
````

````python
# ---- Strip Plot ----
#
# - now let's make a Strip Plot. Realize that the points will occupy
# the same space
#
sns.stripplot(data=insurance_df, x='smoker', y='charges', hue='smoker')

plt.title('Charges by Smoker or Non-Smoker People')
plt.show()
````

<p id='20-bar-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>2.0 | Bar Plots</p>

> **Usage** - `categorical feature frequency`.

`Bar Plots` are used to show the frequency of a Categorical Feature. They are great options to visualize `Frequency and Cross Tables` graphically.

Also, do not misunderstand, frequency means the number of occurences of the feature's values, while distribution shows how the datas of a feature are spread.

> Bar Plot Without Categorical Hue

```python
sns.barplot(data=df, x='Categorical_Feature', y='Frequency')
```

> Bar Plot With Categorical Hue

```python
sns.barplot(data=df, x='Categorical_Feature', y='Frequency', hue='Another_Categorical_Feature')
```

````python
# ---- Bar Plot Without Categorical Hue ----
#
# - using 'value_counts' way
#
sns.barplot(
    data=insurance_df
    , x=insurance_df['smoker'].unique() # ['yes', 'no']
    , y=insurance_df['smoker'].value_counts()
)

plt.title('Frequency of Smoker and Non-Smoker People')
plt.xlabel('Is person a smoker?')
plt.ylabel('Frequency/Count')
plt.show()
````

````python
# ---- Bar Plot Without Categorical Hue ----
#
# - using 'cross_table' way
#
frequencies = pd.crosstab(index=insurance_df['smoker'], columns='count')

sns.barplot(data=frequencies, x=frequencies.index, y='count')

plt.title('Frequency of Smoker and Non-Smoker People')
plt.xlabel('Is person a smoker?')
plt.ylabel('Frequency/Count')
plt.show()
````

````python
# ---- Bar Plot With Categorical Hue ----
#
# - since we are using two numerical features ('bmi' and 'charges')
# in both x and y-axis, it would be better using a scatter plot
# rather than a bar plot.
#
# - however, we did it just for educational purposes!!
#
sns.barplot(
    data=insurance_df[0:20]
    , x='bmi'
    , y='charges'
    , hue='smoker'
)

plt.title('Charges by BMI and Smoker or Not People')
plt.show()
````

<p id='21-histogram-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>2.1 | Histogram Plots</p>

> **Usage** - `numerical feature frequency`.

While Bar Plots are used to show the frequency of Categorical Features, `Histogram Plots` are used to show the frequency of Numerical Features.

> Plot Without Categorical Hue

```python
sns.histplot(data=df, x='Numerical_Feature')
```

> Plot With Categorical Hue

```python
sns.histplot(data=df, x='Numerical_Feature', hue='Categorical_Feature')
```

````python
# ---- Histogram Plot Without Categorical Hue ----
sns.histplot(data=iris_df, x='Sepal Length (cm)')

plt.title('Frequency of Sepal Length')
plt.show()
````

````python
# ---- Histogram Plot With Categorical Hue ----
sns.histplot(data=iris_df, x='Sepal Length (cm)', hue='Species')

plt.title('Frequencies of Sepal Length per Species')
plt.show()
````

````python
# ---- Histogram Plot With Categorical Hue 2 ----
sns.histplot(data=iris_df)

plt.title('Number of Flowers by Sepal Length (cm)')
plt.show()
````

<p id='22-kde-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>2.2 | KDE Plots</p>

> **Usage** - `numerical feature frequency area, being the area between 0 and 1 (Normalized)`.

`KDE Plots` are like `Histograms`, but they do not display the data in bars, but yes, in areas - if that reminds you about Integrals, you are totally in the right road!!

> Plot Without Categorical Hue

```python
sns.kdeplot(data=df, x='Numerical_Feature', fill=True|False)
```

> Plot With Categorical Hue

```python
sns.kdeplot(data=df, x='Numerical_Feature', hue='Categorical_Feature', fill=True|False)
```

````python
# ---- KDE Plot Without Categorical Hue ----
#
# - the 'fill' parameter tells if the are must be painted (True)
# or not (False)
#
sns.kdeplot(data=iris_df, x='Sepal Length (cm)', fill=True)

plt.title('Frequency of Sepal Length')
plt.show()
````

````python
# ---- KDE Plot With Categorical Hue ----
sns.kdeplot(data=iris_df, x='Sepal Length (cm)', hue='Species', fill=True)

plt.title('Frequencies of Sepal Length per Species')
plt.show()
````

````python
# ---- KDE Plot With Categorical Hue 2 ----
sns.kdeplot(data=iris_df, fill=True)

plt.title('Number of Flowers by Sepal Length (cm)')
plt.show()
````

<p id='23-joint-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>2.3 | Joint Plots</p>

> **Usage** - `two numerical features frequency`.

`Joint Plots` union/combine two numerical features frequency, showing their distributions at the x and y-axis's edges and the combination of them in the middle.

This kind of plot is not so good to use due to it be in 3D - unless if you are working with geospatial data.

> Plot Without Categorical Hue
```python
sns.jointplot(data=df, x='Numerical_Feature1', y='Numerical_Feature2', kind='kde', fill=True|False)
```

> Plot With Categorical Hue
```python
sns.jointplot(data=df, x='Numerical_Feature1', y='Numerical_Feature2', kind='kde', fill=True|False, hue='Categorical_Feature')
```

````python
# ---- Joint Plot Without Categorical Hue ----
sns.jointplot(
    data=iris_df
    , x='Sepal Length (cm)'
    , y='Petal Length (cm)'
    , kind='kde'
    , fill=True
)

plt.show()
````

````python
# ---- Joint Plot With Categorical Hue ----
sns.jointplot(
    data=iris_df
    , x='Sepal Length (cm)'
    , y='Petal Length (cm)'
    , hue='Species'
    , kind='kde'
    , fill=True
)

plt.show()
````

<p id='30-heatmap-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>3.0 | HeatMap Plots</p>

> **Usage** - `numerical features correlations`.

`HeatMaps` are especially used to make the correlation table easier to understand, that is, displaying correlations between numerical features graphically.
