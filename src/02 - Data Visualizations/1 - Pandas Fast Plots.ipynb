We have seen how to create plots using Seaborn and Matplotlib so far, now it is time to see how to create with Pandas!!

I called it _Fast Plots_ because we will not have the work to set the plot view, we will just call the function using the DataFrame.

In one side, it is advantegeous because you create plots faster; in other side, it is disadvantegeous due to the settings and styles limitations.

We will see how to create `Bar and Scatter Plots` in this notebook. If you are interesteed to see all available plots in Pandas, check out the [Pandas DataFrame Plot Documentation](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html).

```
- Bar Plots
- Scatter Plots
```

````python
import pandas as pd # pip install pandas
import numpy as np # pip install numpy

np.random.seed(5082023)
````

````python
# ---- Creating Data Frame ----
names = ['Brazil', 'Canada', 'Italy', 'Japan', 'USA']
values1 = [np.random.randint(100), np.random.randint(100), np.random.randint(100), np.random.randint(100), np.random.randint(100)]
values2 = [np.random.randint(100), np.random.randint(100), np.random.randint(100), np.random.randint(100), np.random.randint(100)]

countries_df = pd.DataFrame([names, values1, values2])
countries_df.set_index(pd.Index(['country', 'value_1', 'value_2']), inplace=True)
countries_df = countries_df.T
countries_df
````

<p id='0-bar-plots' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>0 | Bar Plots</p>

For both examples we will be using the `plot` DataFrame function, changing only the parameter `kind` that tells to Pandas what kind of plot we want to use.

````python
# ---- Fast Bar Plots ----
countries_df.plot(
    kind='bar'
    , x='country'
    , y='value_1'
    , legend=False
    , title='Fast Bar Plot Example'
    , xlabel='Country'
    , ylabel='Value 1'
)
````

````python
# ---- Fast Scatter Plot ----
countries_df.plot(
    kind='scatter'
    , x='value_1'
    , y='value_2'
    , legend=False
    , title='Fast Scatter Plot Example'
    , xlabel='Value 1'
    , ylabel='Value 2'
)
````
