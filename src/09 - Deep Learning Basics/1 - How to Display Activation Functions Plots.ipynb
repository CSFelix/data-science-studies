```python
# ---- Importings ----
import matplotlib.pyplot as plt # pip install matplotlib
import mplcyberpunk # pip install mplcyberpunk
import tensorflow as tf # pip install tensorflow
from tensorflow.keras import layers

plt.style.use('cyberpunk')

# ---- Creating Activation Function ----
#
# - try 'linear', 'relu', 'elu', 'selu', 'swish', 'sigmoid' and 'tanh'
#
linear_activation_function = layers.Activation('linear')
relu_activation_function = layers.Activation('relu')
elu_activation_function = layers.Activation('elu')
selu_activation_function = layers.Activation('selu')
swish_activation_function = layers.Activation('swish')
sigmoid_activation_function = layers.Activation('sigmoid')
tanh_activation_function = layers.Activation('tanh')

# ---- Creating Fake Data ----
#
# - tf.linspace(A, B, C): creates a tensor with C equally distributed
# elements from A to B
#
x = tf.linspace(-3.0, 3.0, 100)

linear_y = linear_activation_function(x)
relu_y = relu_activation_function(x)
elu_y = elu_activation_function(x)
selu_y = selu_activation_function(x)
swish_y = swish_activation_function(x)
sigmoid_y = sigmoid_activation_function(x)
tanh_y = tanh_activation_function(x)
```

```python
# ---- Plotting Linear Function ----
plt.figure(dpi=100)
plt.title('Linear Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')

plt.plot(x, linear_y)
plt.xlim(-3, 3)
plt.show()
```

```python
# ---- Plotting Relu Function ----
plt.figure(dpi=100)
plt.title('Relu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')

plt.plot(x, relu_y)
plt.xlim(-3, 3)
plt.show()
```

```python
# ---- Plotting Elu Function ----
plt.figure(dpi=100)
plt.title('Elu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')

plt.plot(x, elu_y)
plt.xlim(-3, 3)
plt.show()
```

```python
# ---- Plotting Selu Function ----
plt.figure(dpi=100)
plt.title('Selu Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')

plt.plot(x, selu_y)
plt.xlim(-3, 3)
plt.show()
```

```python
# ---- Plotting Swish Function ----
plt.figure(dpi=100)
plt.title('Swish Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')

plt.plot(x, swish_y)
plt.xlim(-3, 3)
plt.show()
```

```python
# ---- Plotting Sigmoid Function ----
plt.figure(dpi=100)
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')

plt.plot(x, sigmoid_y)
plt.xlim(-3, 3)
plt.show()
```

```python
# ---- Plotting Tanh Function ----
plt.figure(dpi=100)
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')

plt.plot(x, tanh_y)
plt.xlim(-3, 3)
plt.show()
```
