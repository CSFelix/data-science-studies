The feature extraction performed by the base consists of three basic operations:

> `Filter an image for a particular feature (convolution)`;

> `Detect that feature within the filtered image (ReLU)`;

> `Condense the image to enhance the features (maximum pooling)`.

<h1 id='0-extracting-image-features' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>0 | Extracting Image Features</h1>

```python
# ---- Importations ----
import tensorflow as tf # pip install tensorflow
import matplotlib.pyplot as plt # pip install matplotlib
import mplcyberpunk # pip install mplcyberpunk
import sympy # pip install sympy
from IPython.display import display # pip install IPython

plt.style.use('cyberpunk')

# ---- Reading Image ----
image_path = './datasets/jojo/jojo.jpg'
image = tf.io.read_file(image_path)
image = tf.io.decode_jpeg(image, channels=1) # the channels define the colors. 1 channel: gray scale
image = tf.image.resize(image, size=[400, 400])

plt.figure(figsize=(6, 6))
plt.imshow(tf.squeeze(image), cmap='gray')
plt.axis('off')
plt.show()
```

```python
# ---- Creating the Kernel for the Weights ----
#
# - Edge Detection
#
kernel = tf.constant([
    [-1, -1, -1]
    , [-1, 8, -1]
    , [-1, -1, -1]
])

sympy.init_printing()
display(sympy.Matrix(kernel))
```

```python
# ---- Preparing Image to be Filtered ----
image = tf.image.convert_image_dtype(image, dtype=tf.float32)
image = tf.expand_dims(image, axis=0)

kernel = tf.reshape(kernel, [*kernel.shape, 1, 1])
kernel = tf.cast(kernel, dtype=tf.float32)

# ---- Filtering the Image Features ----
image_filter = tf.nn.conv2d(
    input=image
    , filters=kernel
    , strides=1
    , padding='SAME'
)

plt.figure(figsize=(6,6))
plt.imshow(tf.squeeze(image_filter))
plt.axis('off')
plt.show()
```

```python
# ---- Detecting the Image ----
image_detect = tf.nn.relu(image_filter)

plt.figure(figsize=(6,6))
plt.imshow(tf.squeeze(image_detect))
plt.axis('off')
plt.show()
```
