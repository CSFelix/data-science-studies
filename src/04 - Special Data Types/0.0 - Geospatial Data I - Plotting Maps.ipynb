`GeoPandas (GPD)` is the library to read dataframes that contain Geo Spatial Datas (Geommetry) in the columns. In general, the Spatial Datas can be one of the three below:

> **Point** - `used to represent a location/spot in the map. Attributes: point_gdf.x | point_gdf.y`;

> **Line** - `used to represent streets, rivers, roads and so on. Attributes: line_gdf.length`;

> **Polygon** - `used to represent boundaries. Attributes: polygon_gdf.area`.

For this first part, let's plot all loans in Philippines provided by [Kive.org](https://www.kiva.org).

```python
# ---- Reading the World Dataset Provided by GeoPandas ----
#
# - the 'geometry' variable contains the polygons representing the countries
#
import geopandas as gpd # pip install geopandas
import matplotlib.pyplot as plt # pip install matplotlib
import mplcyberpunk # pip install mplcyberpunk

plt.style.use('cyberpunk')

world_df = gpd.read_file('./datasets/natural_earth/naturalearth_lowres.shp')
world_df.head()
```

```python
# ---- Reading the World Loans Dataset ----
#
# - the 'geometry' variable contains the data points.
#
world_loans_df = gpd.read_file('./datasets/kiva_loans/kiva_loans.shp')
world_loans_df.head()
```

```python
# ---- Plotting all the Loans Around the World ----
#
# - Parameters:
#    \ color: territories' background color;
#    \ edgecolor: territories' edges line color;
#    \ zorder: order to show the territories. It is like 'z-index' attribute in CSS.
#
axis = world_df.plot(
    figsize=(10, 7)
    , color='none'
    , edgecolor='white'
    , linestyle=':'
    , zorder=0
)

world_loans_df.plot(
    color='red'
    , markersize=3
    , ax=axis
    , zorder=1
)

plt.show()
```

```python
# ---- Plotting only Loans on Philippines ----
#
# - let's narrow our plot to only Philippines Territory
#
philippines_world_df = world_df.loc[world_df.name == 'Philippines'].copy()
philippines_loans_df = world_loans_df.loc[world_loans_df.country == 'Philippines'].copy()

axis = philippines_world_df.plot(
    figsize=(10, 7)
    , color='none'
    , edgecolor='white'
    , linestyle=':'
    , zorder=0
)

philippines_loans_df.plot(
    color='red'
    , markersize=3
    , ax=axis
    , zorder=1
)

plt.show()
```
