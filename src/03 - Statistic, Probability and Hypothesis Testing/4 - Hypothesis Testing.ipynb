<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.0 | Hypothesis</h2>

So, our goal in this problem is to determine if the shounen animes average score and the seinen animes average score are equal. For this, we have two conclusions: the averages are equal and the averages are not equal. These two possible conclusions are called **Hypothesis**.

The first Hypothesis is called **Null Hypothesis** that always tell that nothing uncommon is happening, that is, it tells that the averages are equal and both data from shounen and seinen animes came from the same population.

In the other hand, we have the **Alternative Hypothesis** that always tell that something uncommon is happening, that is, it tells that the averages are not equal and both data from shounen and seinen animes came from different populations.

Statistically speaking, we have:

> **Null Hypothesis (H0)** - `anything uncommon is happening because the datas are from the same population`

$$
\text{H0}: \overline X(shounen) = \overline X(seinen)
$$

> **Alternative Hypothesis (H1)** - `something uncommon is happening because the datas are from different populations`

$$
\text{H1}: \overline X(shounen) < \overline X(seinen) \\
\lor \\
\text{H1}: \overline X(shounen) \neq \overline X(seinen) \\
\lor \\
\text{H1}: \overline X(shounen) > \overline X(seinen)
$$

<br />

Having this in mind, the possible outcomes we can achieve after applying the testings are:

> **Fail to Reject the Null Hypothesis (H0)** - `shounen and seinen animes have the same average score and are from the same population`;

> **Reject the Null Hypothesis** - `shounen and seinen animes have different average scores and are from different populations`.

<br />

Now, we are heading to where the Hypothesis Testing becomes interesting and I am going to blow up your mind. Even though we can fail to reject or reject the Null Hypothesis, there is the possibility that our result is wrong, right? For example, when we do a Math exam on school, even though we get a result in our calculations, there is the possibility that the result is wrong and we miss the question; the same thing can happen in Hypothesis Testing. We call them **Errors** and they are bummers ðŸ˜¥

There are two types of Errors and we will learn key concepts and techniques to decrease the possibility of making them. For now, let's see what their types are:

> **Error Type 1 - False Positive** - `the Null Hypothesis is rejected when it should be failed to be rejected. For instance, we concluded that the average score of shounen animes is different from the average score of seinen animes when they are actually equals`;

> **Error Type 2 - False Negative** - `the Null Hypothesis failed to be rejected when it should be rejected. For instance, we concluded that the average score of shounen animes and the average score of seinen animes are equal when they are actually differents`.

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.1 | Level Regions</h2>

Assume that we are talking to the MyAnimeList's owner and he asks us: *"So is the average score of shounen animes equal or different of the average score of seinen animes?"*, and we answer: *"With 100% of certainty and considering that the range of the average scores goes from 0 to 10, the averages are equals!"*. The owner will probably be pissed off on us. Why? Hehe...

Since the scores goes from 0 to 10, when we consider that the average range also goes from 0 to 10, all averages are equal, because their range will always be between 0 and 10, got you?

So, with a 100% of certainty, we can assume that the averages are the same, but since we are taking the minimum and the maximum possible values as the range boundaries, our conclusion is poor and contains no valuable insight. So, instead of having 100% of certainty and making a poor conclusion, what about we allow us to have some uncertainty about the result but make a richer conclusion? Are you going to take the risk? Yes? So do I!!

Normally, **95% (0.95)** is a common acceptable certainty whereas **5% (0.05)** is a common acceptable uncertainty. However, these percentages changes accordingly to the problem and to the risk the client and the Data Scientist decide to handle. For our plot, these percentages are fine!! So, now we are going to give a conclusion that has a chance of 95% (0.95) to be correct and 5% (0.05) to be wrong.

For our conclusion, we have to use the data to calculate the probability of getting a wrong conclusion if we choose to reject the Null Hypothesis when it should be failed to be rejected, that is, the probability of making a **Type 1 Error**. This probability is called **p-value**, goes from 0 to 1 and if it is smaller than the Significance Level, we reject the Null Hypothesis; otherwise, we fail to reject the Null Hypothesis.

In Statistics we call the acceptable certainty as **Confidence Level**, and the acceptable uncertainty as **Significance Level**, being:

> **Confidence Level** - `region/area on a Normal Distribution that shows the probability to the Null Hypothesis be failed to be rejected`;

> **Significance Level** - `region/area on a Normal Distribution that shows the probability to the Null Hypothesis be rejected`;

> **P-Value** - `probability of getting a wrong conclusion if we choose to reject the Null Hypothesis when it should be failed to be rejected. Statistically speaking, it is the probability of making a Type 1 Error`;

>> `When it is smaller than the Significance Level, it means that the probability of making a Type 1 Error is tiny and we can reject the Null Hypothesis. When it is equal or greater than the Significance Level, it means that the probability of making a Type 1 Error is large and we can fail to reject the Null Hypothesis`.

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.2 | Calculating P-Values and Explaining P-Hacking</h2>

Okay, you may be wondering: *"How to calculate the p-value in order to know if the Null Hypothesis must be rejected or fail to be rejected?"*. It depends on the type of Hypothesis Testing we are applying, but all types follow the same logic, they just add up:

1. the probability to observe the desired result in the Distribution - `it tells how rare is the desired result`;
2. the probability of obeservating other results that is equally rare to the desired result - `it tells how many rare events are likely the desired result`;
3. the probability of observating rarer or more extreme results to the desired result - `it tells how many rarer events are to the desired result`.

In a nutshell, the more observations of equally or rarer results, the higher is the p-value and the greater are the odds to fail to reject the Null Hypothesis.

---

Now, I am going to blow up your mind once again, even though the p-value is smaller enough to reject the Null Hypothesis, there are situations we cannot reject ðŸ¤¯. It happens due to an antiethical behavior called **P-Hacking**.

Picture this, consider that we get a sample of shounen and seinen animes, calculate their means, apply a T-Test (do not worry about it now, we will cover it later) for Hypothesis Testing and get a p-value higher than 0.05, then, we can say that the Null Hypothesis failed to be rejected.

But, just in order to get more confidence, we get another sample of shounen and seinen animes, do the same steps and conclude that we failed to reject H0. We do it over and over again and in all experiments get the same conclusion. After taking 429 experiments, we finally took one where the p-value is smaller than the Significance Level. So we concluded that we can finally reject H0, right? Nope! It is wrong!!

Even though in this last experiment we say we can reject H0, we did other 429 experiments comproving the inverse, so what do you think? It would be wrong and antiethical assuming only the result of the last experiment and discarding the result of the other experiments only to get the desired - and manipulated - Null Hypothesis rejection, right? The correct behavior here is to fail to reject the Null Hypothesis!!

I assume you already got the glimpse of how p-hacking works and why it is not cool to do. However, here is another example that you must avoid: we take another sample of shounen and seinen animes and got a p-value equals to 0.06. Since it is pretty close to the Significance Level, we add 2 or 3 more elements into our sample and get a p-value equals to 0.03, concluding that the Null Hypothesis must be rejected. Once again, it is anthietical and the correct behavior is to fail to reject H0.

Once an experiment is done, we cannot change the size and/or the elements of the sample only to have a chance to get the desired result, for instance, to have the chance to reject the Null Hypothesis when it should actually be failed to be rejected. To avoid it, there are techniques to calculate a great sample size for the testings and we will cover how to do it using **Statistical Power**.

> **P-Hacking: Number of Experiments** - `it is not ethical and allowed to take a bunch of experiments until to achieve the desired result. If the major experiments got a p-value higher than the Significance Level and only a few ones got a p-value smaller than, the correct behavior is to fail to reject H0`;

> **P-Hacking: Sample Size** - `also, it is not ethical and allowed to modify the sample size until to achieve the desired result`.

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.3 | Statistical Power and Power Analysis</h2>

Besides the **P-Hacking**, we have to deal with another problem: the situation where we failed to reject the Null Hypothesis when it should be rejected instead - **Error Type 2**. But do not worry, because **Statistical Power** and **Power Analysis** are here to solve both of them for us.

Kicking off with **Statistical Power**, it is the probability that we will correctly reject the Null Hypothesis, being **80% (0.80)** a common used value. Then, if we take a Hypothesis Testing and the choosen Statistical Power is 80%, we are literally saying that: *"If we conclude rejecting the Null Hypothesis, we have 80% of probability that our conclusion is correct."*. It may be kind of confusing when you think about it and the Confidence Level, but relax, I will wrap everything up in the end and all your doubts will disappear!!

So, just stick into your mind that Statistical Power is the probability of correct rejecting the Null Hypothesis and 80% is the standard value for many cases.

---

However, only determining that the Statistical Power is 80% nothing will change, we will still have to deal with p-hacking and the Error Type 2. So we use the **Power Analysis** in order to put the Statistical Power into action!!

With Power Analysis we determine a great size for our samples before taking the testings using the Statistical Power as a parameter. So at the same time we avoid p-hacking, we also avoid Error Type 2, killing two birds with one stone.

To calculate the Power Analysis for sample size, we have to follow four steps:

1. Determine the Power: `80% (0.80)`;

2. Determine the Significance Level: `5% (0.05)`;

3. Estimate the Overlap between the distributions;

4. Calculate the Sample Size accordingly to the Hypothesis Test to be Applied.

<h1 id='1-preparing-the-tests' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>1 | Preparing the Tests</h1>

Now, we are going to prepare the Hypothesis Testings using the animes dataset!! Our problem here is to determine the the average rating of action animes and the average rating of romance animes are equal or not. We have choosen these two genres since they are very different between them, while action focuses on fights and super powers, romance focuses on love and real life; and, consequently, we can compare their audience ratings.

Defined our problem, the next step is to choose what Hypothesis Testing we are going to use. For this scenario, T-Test or Z-Test works perfectly - do not worry, we will cover them later explaining how they works and when to use which one.

Going on, let's consider the Confidence Level as 95% and the Significance Level as 5%. Also, in order to avoid p-hacking, we are going to use samples to give us 80% of Power.

So far, we have:

> **Testing Technique** - `T-Test or Z-Test`;

<br />

> **Null Hypothesis** - $\text{H0}: \overline X(action) = \overline X(romance)$;

> **Alternative Hypothesis** - $\text{H1}: \overline X(action) \neq \overline X(romance)$;

<br />

> **Confidence Level** - `95% (0.95)`;

> **Significance Level** - `5% (0.05)`;

<br />

> **Statistical Power** - `80% (0.80)`;

> **Samples Size** - `?`.

<br />

Before going on, there are two things to put in mind: 

1. in order to apply T-Test or Z-Test, our data must be Normal Distributed;

2. in order to assure 80% of Power, we must calculate the Power Analysis to find the great sample size.

For the first one, we can use Kolmogorov-Smirnov (KS) Test or Shapiro Wilks Test to check out normality and, for the second one, we follow the steps to calculate the Power Analysis after assuring that the data is Normal Distributed.

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>1.0 | Checking Normality</h2>

First, let's prepare the Python Environment:

````python
# ---- Imports ----
from diptest import diptest                 # pip install diptest
import matplotlib.pyplot as plt             # pip install matplotlib
import mplcyberpunk                         # pip install mplcyberpunk
import numpy as np                          # pip install numpy
import pandas as pd                         # pip install pandas
import seaborn as sns                       # pip install seaborn
import statsmodels.stats.power as smp       # pip install statsmodels
import scipy.stats as stats                 # pip install scipy

from math import sqrt                       # pip install math

# ---- Constants ----
ANIMES_GENRES = ('genre_action', 'genre_romance')
DATASETS_PATH = ('./datasets')
SEED = (20231124)

SIGNIFICANCE_LEVEL = (0.05)
STATISTICAL_POWER = (0.80)

# ---- Settings ----
np.random.seed(SEED)
pd.set_option('display.max_columns', None)
sns.set_style('darkgrid')
plt.style.use('cyberpunk')

# ---- Functions ----
def monte_carlo_simulation_clt(dataset, number_elements=30, experiments=1000):
    """
    \ Description:
        - Creates a list with "experiments" samples with "number_elements" elements - all randomly choosen;
        - Returns the list.
    
    \ Parameters:
        - dataset: pandas DataFrame or pandas Series;
        - number_elements: integer;
        - experiments: integer.
    """
    samples_mean_list = []
    
    for experiment_id in range(experiments):
        np.random.seed(SEED + experiment_id)
        indexes = np.random.randint(0, dataset.shape[0], number_elements)
        samples_mean_list.append(dataset.iloc[indexes].mean())
        
    return samples_mean_list

def cohen_effect_size(dataset_a, dataset_b):
    """
    \ Description:
        - Calculates the Estimated Difference between the Means between both parameters;
        - Calculates the Pooled Estimated Standard Deviations between both parameters;
        - Uses the previous results to calculate the Cohen's Effect Size;
        - Returns Cohen's Effect Size
    
    \ Parameters:
        - dataset_a: list;
        - dataset_b: list.
    """
    estimated_difference_means = np.mean(dataset_a) - np.mean(dataset_b)
    pooled_estimated_std = sqrt((np.std(dataset_a)**2 + np.std(dataset_b)**2) / 2)
    return estimated_difference_means / pooled_estimated_std
````

````python
# ---- Reading Dataset and Filtering Animes ----
full_animes_df = pd.read_csv(f'./{DATASETS_PATH}/anime-exploratory-dataset-2023.csv')

action_animes_df = full_animes_df.loc[
    (full_animes_df[ANIMES_GENRES[0]] == 1) & (full_animes_df['score'] != -1)
]['score'].copy()

romance_animes_df = full_animes_df.loc[
    (full_animes_df[ANIMES_GENRES[1]] == 1) & (full_animes_df['score'] != -1)
]['score'].copy()

action_animes_percentage = round(action_animes_df.shape[0] * 100 / full_animes_df.shape[0], 4)
romance_animes_percentage = round(romance_animes_df.shape[0] * 100 / full_animes_df.shape[0], 4)

print(f'Total Number of Animes: {full_animes_df.shape[0]} - 100%')
print(f'Total Number of Action Animes: {action_animes_df.shape[0]} - {action_animes_percentage}%')
print(f'Total Number of Romance Animes: {romance_animes_df.shape[0]} - {romance_animes_percentage}%')
````

Now, let's check out the Normality using `Kolmogorov-Smirnov (KS) Test` and `Shapiro Wilks Test`. In order to know when to use which one, we follow the rules:

> **KS Test** - `sample size greater than 30 - n > 30`;

> **Shapiro Test** - `sample size between 4 and 2000 = 4 <= n <= 2000`.

<br />

Since action animes sample contains 3,957 observations, there is no problem in using KS Test. In the other hand, there are only 1,831 observations for romance animes, so, for this one, we are going to apply Shapiro Test. Also, to make the testings' results even more powerfull, we will plot the distributions!!

For both tests, these are our Hypothesis:

> **Null Hypothesis** - `the sample's distribution is normal`;

> **Alternative Hypothesis** - `the sample's distribution is not normal`.

````python
# ---- Checking Normality: Action Animes with KS Test ----
#
# \ H0: the sample's distribution is normal;
# \ H1: the sample's distribution is not normal.
#
ks_result = stats.kstest(action_animes_df, 'norm')
print(ks_result)

# ---- Distribution Plot ----
plt.title('Action Animes Score - Distribution')
plt.xlabel('Score')
plt.ylabel('Density')

ax = sns.histplot(action_animes_df, kde=True, bins=30, color='#7159c1')
ax.lines[0].set_color('#bd00ff')

plt.show()
````

````python
# ---- Checking Normality: Romance Animes with Shapiro Test ----
#
# \ H0: the sample's distribution is normal;
# \ h1: the sample's distribution is not normal.
#
shapiro_result = stats.shapiro(romance_animes_df)
print(shapiro_result)

# ---- Distribution Plot ----
plt.title('Romance Animes Score - Distribution')
plt.xlabel('Score')
plt.ylabel('Density')

ax = sns.histplot(romance_animes_df, kde=True, bins=30, color='#7159c1')
ax.lines[0].set_color('#bd00ff')

plt.show()
````

Uh-oh, accordingly the p-values, we can conclude that both distributions are not Normal and, looking at the histograms, we can say that they look like bimodal distributions. Let's take a Dip Test just to assure that our guess is correct.

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>1.1 Checking Bimodality</h2>

These are our new Hypothesis:

> **Null Hypothesis** - `the sample's distribution is bimodal`;

> **Alternative Hypothesis** - `the sample's distribution is not bimodal`.

````python
# ---- Checking Bimodality: Action and Romance Animes with Dip Test ----***
action_animes_dip_statistic, action_animes_dip_p_value = diptest(action_animes_df)
romance_animes_dip_statistic, romance_animes_dip_p_value = diptest(romance_animes_df)

print(f'Action Animes - Statistic: {action_animes_dip_statistic} - P-Value: {action_animes_dip_p_value}')
print(f'Romance Animes - Statistic: {romance_animes_dip_statistic} - P-Value: {romance_animes_dip_p_value}')
````

We got it!! Both p-values are way higher than the Significance Level (0.05) and, using the distributions plots as a parameter, we can securitly affirm that both distributions are Bimodal!!

Okay, we have just proven that the distributions are bimodal, but they must be normal in order to apply Z-Test and T-Test. So, is all our Hypothesis Testing gone? Nope!! We can fix it using the Statistical Knowledge of Central Limit Theorem!!

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>1.2 | Central Limit Theorem (CLT)</h2>

The Central Limit Theorem is a rule in Statistic that says: *"No matter how the distribution is, it can always be transformed into a Normal Distribution. The unique exception is the Cauchy Distribution"*. So, once again, the Statistic wins!!

The steps to transform the distributions into normal are few and easy:

1. Collect a sample of the dataset;
2. Calculate its mean;
3. Plot the mean bar;
4. Go to step 1 and repeat it many times;
5. After repeating the process a certain amount of times, the plotted bar plots will follow a Normal Distribution;
6. Use this Normal Distribution of the means from now on.

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>1.3 | Applying Central Limit Theorem (CLT)</h2>

Now, let's prove the CLT really works. For this, let's take 5, 10, 100 and 1000 samples with 30 elements from action animes and plot them. If the distribution tends to be Normal, we are going to do a Monte Carlo Simulation taking 1,000 samples of action and romance animes and use them to keep going on our experiments.

````python
# ---- Applying CLT ----
number_of_elements = 30
number_of_samples = 1000
samples_mean_list = []

for sample in range(number_of_samples):
    np.random.seed(SEED + sample)
    indexes = np.random.randint(0, action_animes_df.shape[0], number_of_elements)
    samples_mean_list.append(action_animes_df.iloc[indexes].mean())
````

````python
# ---- Plotting Distributions ----
samples_mean_size = [5, 10, 100, 1000]
figure, axis = plt.subplots(2, 2, figsize=(7, 9))

for row_index in range(0, 2):
    for column_index in range(0, 2):
        current_sample_mean_size = samples_mean_size[row_index * 2 + column_index]
        
        # Shapiro Test #
        shapiro_result = stats.shapiro(samples_mean_list[:current_sample_mean_size])
        print(f'{current_sample_mean_size} Samples - {shapiro_result}')
        
        # Plotting #
        axis[row_index, column_index].set_title(f'Means Distribution - {current_sample_mean_size} Sample(s)')        
        sns.histplot(samples_mean_list[:current_sample_mean_size], kde=True, bins=30, color='#7159c1', ax=axis[row_index, column_index])
        
        # Changing KDE Line Color #
        kde_lines = axis[row_index, column_index].get_lines()
        if len(kde_lines) > 0: kde_lines[0].set_color('#bd00ff')
        
plt.show()
````

Woow, realize that the more samples the distribution has, the more it looks like a Normal Distribution. Also, all Shapiro Tests gave us a p-value higher than the Significance Level, comproving that the distributions are Normal!!

Now, let's do Monte Carlo Simulations in order to get 1,000 samples for action and romance animes, check out their normality with Shapiro Test andplot their distributions.

````python
# ---- Monte Carlo Simulation ----
action_animes_samples_mean = monte_carlo_simulation_clt(action_animes_df)
romance_animes_samples_mean = monte_carlo_simulation_clt(romance_animes_df)
````

````python
# ---- Checking out for Normality: Action Animes Samples Means ----
shapiro_result = stats.shapiro(action_animes_samples_mean)
print(shapiro_result)

# ---- Distribution Plot ----
plt.title('Action Animes Samples Mean Score - Distribution')
plt.xlabel('Score')
plt.ylabel('Density')

ax = sns.histplot(action_animes_samples_mean, kde=True, bins=30, color='#7159c1')
ax.lines[0].set_color('#bd00ff')

plt.show()
````

````python
# ---- Checking out for Normality: Romance Animes Samples Means ----
shapiro_result = stats.shapiro(romance_animes_samples_mean)
print(shapiro_result)

# ---- Distribution Plot ----
plt.title('Romance Animes Samples Mean Score - Distribution')
plt.xlabel('Score')
plt.ylabel('Density')

ax = sns.histplot(romance_animes_samples_mean, kde=True, bins=30, color='#7159c1')
ax.lines[0].set_color('#bd00ff')

plt.show()
````

Hoorrayy, our distributions look like Normal!! ðŸŽ‰

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>1.4 | Applying Power Analysis</h2>

However, does a 1,000 samples give us a Statistical Power of 80% in order to avoid the Error Type 2 and p-hacking when taking the Z-Test and T-Test? Let's figure it out calculating the Statistical Analisys!!

````python
# ---- Calculating Great Sample Size ----
effect_size = cohen_effect_size(action_animes_samples_mean, romance_animes_samples_mean)

power_analysis = smp.TTestIndPower()

sample_size = round(power_analysis.solve_power(
    effect_size=effect_size
    , power=STATISTICAL_POWER
    , alpha=SIGNIFICANCE_LEVEL
))

print(f'To get {STATISTICAL_POWER * 100}% of Power with {SIGNIFICANCE_LEVEL} of Signficance Level, it is needed samples with {sample_size} elements at least.')
````

We got that 20 elements is enough to give us 80% of Power and, since action animes and romance animes datasets contain 1,000 elements each, the data is almost ready for the final Hypothesis Testing!!

<h2 style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>1.5 | Checking Variance Homogeneity</h2>

There is only one more thing to assure before going to the last test: check out if the variance of the datasets is homogeneity, that is, if their variance are equal or not. It is needed due to a simple parameter from T-Test and Z-Test called `equal_var`. For this step, we are going to use `F-Levene Test`.

> **Null Hypothesis** - `action animes and romance animes have the same variance`;

> **Alternative Hypothesis** - `action animes and romance animes does not have the same variance`.

````python
# ---- Checking Variance Homogeneity: F-Levene ----
stats.levene(action_animes_samples_mean, romance_animes_samples_mean, center='mean')
````

Woow, the p-valuee is way smaller than the Significance Level (0.05), so the variances of both datasets are not equal!!

<h1 id='2-testing-on-anime-dataset' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>2 | Testing on Anime Datasets</h1>

Okay, so far, we have done:

1. Transformed both action and romance animes datasets into a Normal Distribution applying the Central Limit Theorem (CLT);
2. and Calculated the Power Analysis for 80% of Power.

Now it is time for the final goal: check out whether or not action and romance animes average score are equal. To do this, we have two options of Hypothesis Testing: **Z-Test and T-Test**.

The difference between them is easy to catch, while Z-Test uses the population's standard deviation, T-Test uses the sample's standard deviation in their equations:

$$
\text{Z-Test}: \frac{\overline X - \mu} {\frac{\sigma} {\sqrt{n}}}
$$

$$
\text{T-Test}: \frac{\overline X - \mu} {\frac{S} {\sqrt{n}}}
$$

where:

$$
\overline X : \text{sample's mean}
$$

$$
\mu : \text{population's mean}
$$

$$
n : \text{sample's size}
$$

$$
\sigma : \text{population's standard deviation}
$$

$$
S : \text{sample's standard deviation}
$$

---

There is a rule of thumb to choose which test to use: when we know the standard deviation of the population and the sample size is larger than 30 elements, we use `Z-Test`; on the other hand, when we do not know the population's standard deviation or the sample size is smaller than or equal to 30 elements, we use `T-Test`.

Even though both animes datasets has more than 30 elements and the population's standard deviation is known - it is the full animes dataset's standard deviation -, we are going to use `T-Test`. The reason? Simple, because we are afraid to make mistakes assuming that both datasets are really similar to the population. Yeah, the reason here are just fo security.

So, let's apply the final Test!!

> **Null Hypothesis** - `action animes and romance animes have the same average score`;

> **Alternative Hypothesis** - `action animes and romance animes does not have the same average score`.

````python
# ---- T-Test ----
stats.ttest_ind(
    a=action_animes_samples_mean
    , b=romance_animes_samples_mean
    , alternative='two-sided' # H1: mean(a) != mean(b)
    , equal_var=False
    , random_state=SEED
)
````

The p-value for T-Test is also way smaller than the Significance Level, so we can assume that action animes and romance animes have different average scores and, consequently, that their public are different!!
