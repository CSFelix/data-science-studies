```python
# ---- Settings ----
import pandas as pd # pip install pandas
import numpy as np # pip install numpy

np.random.seed(20241801)
```

```python
# ---- Reading Dataset ----
catalog_df = pd.read_csv('./datasets/catalog.csv')

# ---- Checking Existence of Duplicated Values ----
#
#	\ False >> there are no duplicated values
#	\ True  >> there are duplicated values
#
catalog_df.duplicated('country_name').any()
```

```python
# ---- Creating Duplicated DataFrame ----
duplicated_catalog_df = catalog_df[catalog_df.duplicated('country_name')]
duplicated_catalog_df.head()
```
